```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
import pandas as pd
```

```python
file = 'W_C.xlsx'
data = pd.read_excel(file)
```

```python
data
```

```python
df = pd.DataFrame(data)
df['HOURS']=pd.date_range(start='2022-10-30 12',end='2022-10-31 12',freq='3H')
df = df.set_index('HOURS')
```

```python
df
```

```python
T = df['TEMP']
R = df['RH']
```

```python
fig, axes = plt.subplots(figsize=(8,5))


ax1=axes
ax1.plot(df.index,T,color='red',lw=1.5,ls='-',marker='o',label='TEMPERATURE')
ax1.set_xlabel('TIME(UTC)',fontsize=15)
ax1.set_ylabel('TEMPERATURE($^oc$)',color='red',fontsize=15)

ax2=axes.twinx()
ax2.plot(df.index,R,color='blue',lw=1.5,ls='-',marker='*',label='RELATIVE HUMIDITY')
ax2.set_ylabel('RELATIVE HUMIDITY(%)',color='blue',fontsize=15)
ax2.set_title('TEMPERATURE VS RELATIVE HUMIDITY',fontsize=15)
fig.savefig('temp.jpeg')
```
